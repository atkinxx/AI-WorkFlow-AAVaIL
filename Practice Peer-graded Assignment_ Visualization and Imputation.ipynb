```python
import pandas
dataset = pandas.read_csv('drive/My Drive/IBM Course/AAVAIL_task.csv')
dataset.head()
```

```python
country_subscriber = dataset[['country_name', 'is_subscriber']]
def sub(x):
  if x==True:
    return 1
  else:
    return 0
country_subscriber['is_subscriber'] = country_subscriber['is_subscriber'].apply(sub)
country_subscriber.head()
```

```python
us = country_subscriber[country_subscriber['country_name']=='united_states']
singapore = country_subscriber[country_subscriber['country_name']=='singapore']

us = us['is_subscriber'].sum()
singapore = singapore['is_subscriber'].sum()

print(us)
print(singapore)
```

```python
import matplotlib.pyplot as plt
plt.bar(country_subscriber['country_name'].unique(), [us,singapore])
plt.show()
```
